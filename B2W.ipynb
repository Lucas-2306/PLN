```python
import pandas as pd

df = pd.read_csv("Docs/B2W-Reviews01.csv")

df = df[['overall_rating', 'review_text']]

df = df.dropna(subset=['review_text'])

df = df.drop_duplicates(subset=['review_text'])

def classify_review(row):
    rating = row['overall_rating']
    
    if rating >= 4:
        return 'Positivo'
    elif rating <= 2:
        return 'Negativo'
    else:
        return 'Neutro'

df['Classificacao'] = df.apply(classify_review, axis=1)

df = df.drop(["overall_rating"], axis = 1)

df.rename(columns={'review_text': 'Text'}, inplace=True)

Q1 = 68
Q3 = 154
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Removing Outliers

df = df[(df['Text'].str.len() <= upper_bound) & (df['Text'].str.len() >= lower_bound)]

print(df)
print(len(df[df['Classificacao'] == "Positivo"]))
print(len(df[df['Classificacao'] == "Negativo"]))
print(len(df[df['Classificacao'] == "Neutro"]))
#print(df['Text'].str.len().describe())

df.to_csv("B2W.csv")
```
