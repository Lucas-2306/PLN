```python
from classificador import Classificador
from Modules.utils import progressBar
import nltk
import pandas as pd
import joblib

classificador = Classificador()

#resposta = lexico.classic_classification("Matou uma criança")
#print(resposta)

df = pd.read_csv("B2W.csv")
#print(df)

total = len(df)
hits = 0
hits_e = 0
hits_n = 0
hits_p = 0

#total = 100 #O dataframe inteiro demora bastante
inicio = 0
step = 1

clf = joblib.load('sentiment_model.pkl')
vectorizer = joblib.load('tfidf_vectorizer.pkl')

for i in progressBar(range(inicio, inicio + total, step), prefix = 'Progress:', suffix = 'Complete', length = 50):
    classification = classificador.model_classification(df['Text'][i], clf, vectorizer)
    #classification = classificador.pretrained_model_classification(df['Text'])

    #print(classification)

    #print(df['Classificacao'][i])

    if(classification == df['Classificacao'][i]):
        if(classification == "Negativo"):
            hits_n += 1
        if(classification == "Positivo"):
            hits_p += 1
        if(classification == "Neutro"):
            hits_e += 1
        hits += 1

accuracy = hits / (total / step)
print(hits)
print(hits_p)
print(hits_n)
print(hits_e)
print(accuracy)
```
